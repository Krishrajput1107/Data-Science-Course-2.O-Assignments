1.Histogram:

Use: Show data distribution.
Function: sns.histplot()

Scatter Plot:

Use: Display variable relationships.
Function: sns.scatterplot()

Bar Plot:

Use: Compare categories or frequencies.
Function: sns.barplot()

Pair Plot:

Use: Explore variable pairs.
Function: sns.pairplot()

Box Plot:

Use: Visualize distribution and outliers.
Function: sns.boxplot()

Seaborn simplifies these visualizations with its easy-to-use functions and aesthetically pleasing styles.







```python
#2.
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

plt.title("FMRI Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#3.
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
#4.
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack')

plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.legend(title='Cut')
plt.tight_layout()
plt.show()
```

```python
#5.
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = sns.load_dataset("iris")

sns.set(style="ticks")
sns.pairplot(iris_data, hue="species")

plt.title('Pair Plot of Iris Dataset by Species')
plt.show()
```
