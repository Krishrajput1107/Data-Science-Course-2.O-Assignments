Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Answer: Estimation statistics involves using sample data to make inferences or estimates about population parameters. There are two main types of estimation:

Point Estimate: A point estimate is a single value that is used to estimate a population parameter. For example, the sample mean is often used as a point estimate for the population mean.

Interval Estimate: An interval estimate provides a range of values within which the population parameter is likely to fall. This range is typically expressed as a confidence interval. Confidence intervals are used to quantify the uncertainty associated with point estimates.

Q2: Write a Python function to estimate the population mean using a sample mean and standard deviation.

Answer:

```python
import numpy as np
import scipy.stats as stats

def estimate_population_mean(sample_mean, sample_std, sample_size):
    """
    Estimate the population mean using a sample mean, standard deviation, and sample size.

    Parameters:
    sample_mean (float): Sample mean.
    sample_std (float): Sample standard deviation.
    sample_size (int): Sample size.

    Returns:
    population_mean_estimate (float): Estimated population mean.
    """
    standard_error = sample_std / np.sqrt(sample_size)
    margin_of_error = stats.norm.ppf(0.975) * standard_error  # 95% confidence interval
    population_mean_estimate = sample_mean - margin_of_error
    return population_mean_estimate
```

Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

Answer: Hypothesis testing is a statistical method used to make inferences about population parameters based on sample data. It involves formulating a null hypothesis (H0) and an alternative hypothesis (Ha), collecting data, and using statistical tests to determine whether there is enough evidence to reject the null hypothesis.

Importance of Hypothesis Testing:

Hypothesis testing is essential for making data-driven decisions in various fields, including science, business, and healthcare.
It helps assess the significance of observed effects, draw conclusions, and make predictions.
It provides a systematic and objective way to evaluate hypotheses and claims based on empirical evidence.

Q5: Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

Answer:

```python
import scipy.stats as stats

def hypothesis_test_two_means(sample1, sample2, alpha):
    """
    Conduct a hypothesis test on the difference between two population means.

    Parameters:
    sample1 (array-like): Sample data for the first group.
    sample2 (array-like): Sample data for the second group.
    alpha (float): Significance level.

    Returns:
    result (str): Result of the hypothesis test (e.g., "Reject H0" or "Fail to reject H0").
    p_value (float): p-value of the test.
    """
    stat, p_value = stats.ttest_ind(sample1, sample2)
    if p_value < alpha:
        result = "Reject H0"
    else:
        result = "Fail to reject H0"
    return result, p_value
```

Answer:

Null Hypothesis (H0): A statement of no effect, no difference, or no relationship. It represents the status quo or a baseline assumption.

Alternative Hypothesis (Ha): A statement that contradicts the null hypothesis, typically suggesting an effect, difference, or relationship of interest.

Examples:

Alternative Hypothesis: The new drug decreases blood pressure.

Q7: Write down the steps involved in hypothesis testing.

Answer:
Steps in Hypothesis Testing:

Formulate the null hypothesis (H0) and the alternative hypothesis (Ha).
Collect sample data.
Choose a significance level (alpha) to determine the threshold for significance.
Perform a statistical test using appropriate methods (e.g., t-test, z-test).
Calculate the test statistic and p-value.
Compare the p-value to the significance level.
Make a decision: "Reject H0" if p-value < alpha, or "Fail to reject H0" if p-value >= alpha.
Draw conclusions and interpret the results.

Q8: Define p-value and explain its significance in hypothesis testing.

Answer: The p-value is the probability of obtaining test results as extreme or more extreme than the observed results, assuming the null hypothesis is true. It quantifies the evidence against the null hypothesis. In hypothesis testing:

A small p-value (typically less than the chosen significance level, alpha) indicates strong evidence against the null hypothesis, leading to its rejection.
A large p-value suggests weak evidence against the null hypothesis, leading to a failure to reject the null hypothesis.
The p-value helps determine whether the observed results are statistically significant and whether there is enough evidence to support the alternative hypothesis.


Q9: Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data points for a t-distribution with 10 degrees of freedom
degrees_of_freedom = 10
x = np.linspace(-4, 4, 1000)
pdf = stats.t.pdf(x, df=degrees_of_freedom)

# Plot the t-distribution
plt.plot(x, pdf, label=f't-distribution (df={degrees_of_freedom})')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Student\'s t-Distribution')
plt.show()
```

Q10: Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

Answer:



```python
import scipy.stats as stats

def two_sample_t_test(sample1, sample2, alpha):
    """
    Conduct a two-sample t-test for independent samples with equal variances.

    Parameters:
    sample1 (array-like): Sample data for the first group.
    sample2 (array-like): Sample data for the second group.
    alpha (float): Significance level.

    Returns:
    result (str): Result of the hypothesis test (e.g., "Reject H0" or "Fail to reject H0").
    p_value (float): p-value of the test.
    """
    stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)
    if p_value < alpha:
        result = "Reject H0"
    else:
        result = "Fail to reject H0"
    return result, p_value
```

Q11: What is Student’s t distribution? When to use the t-Distribution.

Answer: The Student's t-distribution, often referred to as the t-distribution, is a probability distribution used in statistical hypothesis testing and parameter estimation. It is characterized by its bell-shaped curve, like the normal distribution, but has heavier tails. The t-distribution is used in the following scenarios:

When dealing with small sample sizes (typically n < 30) where the population standard deviation is unknown.
When estimating population parameters (e.g., mean) from a sample and quantifying the uncertainty associated with the estimate.
In cases where the data are approximately normally distributed or if the sample size is large, the t-distribution approaches the standard normal distribution.
