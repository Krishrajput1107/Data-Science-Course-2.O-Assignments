Q1: The main difference between a t-test and a z-test is in the scenarios in which they are used and the assumptions they make.

T-test: A t-test is used when you want to compare the means of two groups or samples, and you do not know the population standard deviation. It assumes that the data follows a normal distribution or is approximately normally distributed. For example, you might use a t-test to compare the average test scores of two different classes of students.

Z-test: A z-test is used when you want to compare a sample mean to a known population mean and you know the population standard deviation. It assumes that the data follows a normal distribution or is approximately normally distributed. For instance, you might use a z-test to determine if the average height of a sample of people is significantly different from the known average height of the entire population.

Q2:

One-tailed test: In a one-tailed test, you are testing a specific direction of an effect. For example, you might be interested in whether a new drug increases the average lifespan (specifically longer), so you conduct a one-tailed test looking for an effect in that direction.

Two-tailed test: In a two-tailed test, you are testing for any significant difference without specifying a direction. For example, you might want to determine if a coin is biased (either it comes up heads more often or tails more often) compared to an unbiased coin.

Q3:

Type 1 error: This occurs when you reject a true null hypothesis. In other words, you conclude that there is an effect or difference when there isn't. An example could be a medical test falsely indicating a person has a disease when they do not.

Type 2 error: This occurs when you fail to reject a false null hypothesis. In other words, you conclude that there is no effect or difference when there actually is. An example could be a medical test failing to detect a disease when the person is actually sick.

Q8: The margin of error in a confidence interval is a measure of the uncertainty associated with the estimate. It depends on the confidence level and the standard error of the estimate. A larger sample size typically results in a smaller margin of error because larger samples provide more information and reduce sampling variability.

Example: Suppose you want to estimate the average income of households in a city. If you take a small random sample of 20 households, the margin of error might be $1,000 at a 95% confidence level. If you increase the sample size to 200 households, the margin of error might decrease to $100 at the same confidence level, providing a more precise estimate.
